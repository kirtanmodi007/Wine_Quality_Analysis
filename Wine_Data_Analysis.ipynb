<h2> Now we want to append both red_df (Red Wine DataFrame) and white_df (White Wine DataFrame) into one dataframe wine_df

<h4>  <ul>Before appending both datarframe into one dataframe we have to add one extra column in both dataframe which reflects the type of wine.</ul>

<ul>Inorder to add one column into Particular Dataframe we have to add array of the size of dataframe with Value of that particular wine type..


```python
import numpy as np
```

```python
#This belowed line of code will general a color_red array with RED value repeated for the size of that dataframe.
#In our case there are 1599 rows into red_df dataframe. So, our array will have "red" value for 1599 times.
color_red = np.repeat("red",red_df.shape[0])
```

<ul> After making the array we will simply add that array as a New column in red_df dataframe.

```python
red_df['color'] = color_red
```

```python
red_df
```

<h3> We will repeat the same process for white_wine_data to add one extra column.

```python
white_color = np.repeat("white", white_df.shape[0])
```

```python
white_df['color'] = white_color
```

```python
white_df
```

<h2> Now we have 1 extra column added into both dataframe. Now we just have to append both the dataframe into one new dataframe. WHich will have combined column of both the dataset.

```python
wine_df = red_df.append(white_df, ignore_index = True) 
```

```python
wine_df
```

<h2> So, now we have combined dataset with wine color into one dataframe named wine_df.

```python
#BY doing this we can remove all the rows from wine_df dataframe.
wine_df = wine_df[0:0]
```
